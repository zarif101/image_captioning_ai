```python
def clean_caption(x):
    cap = x.split()
    cap = [w.lower() for w in cap]
    cap = [w for w in cap if w not in stopwords.words('english')]
    cap = [w for w in cap if w not in string.punctuation]
    return ' '.join(cap)
annots['cleaned_caption'] = annots['caption'].apply(clean_caption)
```

```python
annots.head(3)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
def reduce_count(st):
    words = st.split()
    chars = ['-','.','&','=']
    for char in string.punctuation:
        words = [w for w in words if char not in w]
    words = [w for w in words if 5000 > tokenizer.word_counts[w] > 10]
    words.insert(0,'startseq')
    words.append('endseq')
    words = ' '.join(words)
    
    return words

for i in range(2):
    tokenizer = Tokenizer(oov_token='OOV')
    tokenizer.fit_on_texts(annots['cleaned_caption'])
    annots['cleaned_caption'] = annots['cleaned_caption'].apply(reduce_count)
    tokenizer = Tokenizer(oov_token='OOV')
    tokenizer.fit_on_texts(annots['cleaned_caption'])
```

```python
max(tokenizer.word_index.values())
```

```python
tups_list = []
```

```python
def gen_tups():
    global tups_list
    for ix,row in annots.iterrows():
        imname = row[0]
        whole_caption = row[2]
        caption = whole_caption.split(' ')
        caption = [z for z in caption if z not in ['.',',']]
        path = 'data/flickr8k-sau/Flickr_Data/Images/'+imname.split('#')[0]
        for i in range(1,len(caption)):
            cap = caption[:i]
            cap = [' '.join(cap)]
            target = [caption[i]]
           
            caption_seq = tokenizer.texts_to_sequences(cap)
            
            target_seq = tokenizer.texts_to_sequences(target)
            caption_seq = pad_sequences(caption_seq,maxlen=10)[0]
           
            for x in caption_seq:
                if x == 1:
                    print(caption_seq)
            caption_seq = np.array(caption_seq)
            target_seq = np.array(target_seq)
            
            tup = (path,caption_seq,target_seq) #path to img, caption so far, target word
            tups_list.append(tup)
gen_tups()
```

```python
feat_model = keras.applications.vgg16.VGG16(weights='imagenet')
```

```python
feat_model = Model(feat_model.input,feat_model.layers[-2].output)
```

```python
test_img = cv2.imread('data/flickr8k-sau/Flickr_Data/Images/1000268201_693b08cb0e.jpg')
test_img = cv2.resize(test_img,(224,224)).reshape((1,224,224,-1))
feat_model(test_img).shape
```

```python
def Generator(tups_list,batch_size):
    while 1:
        captions = []
        images = []
        targets = []
        batch = random.choices(tups_list,k=batch_size)
        idx = 1
        for each in batch:
            try:
                path = each[0]
                caption = each[1]
                target = each[2]

                captions.append(caption)

                target = to_categorical(target,num_classes=1754)

                targets.append(target)

                image = cv2.imread(path)
                image = cv2.resize(image,(224,224)).reshape((1,224,224,-1))
                image = np.array(feat_model(image)).reshape((4096,))
                images.append(image)
                idx+=1
            except:
                print('ERROR!')
                captions = []
                images = []
                targets = []
                batch = random.choices(tups_list,k=batch_size)
        #print(np.array(captions))
        yield [np.array(images),np.array(captions).reshape((-1,10))],np.array(targets).reshape((-1,1754))
```

```python
from tensorflow.keras.models import Model,load_model
import tensorflow
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Embedding,LSTM,Dense,Conv2D,MaxPool2D,Dropout,Input,add,Flatten
```

```python
seq_i = Input(shape=(10,))
seq = Embedding(1755,200,input_length=10,mask_zero=True)(seq_i)
seq = Dropout(rate=0.2)(seq)
seq = LSTM(units=256)(seq)

pic_i = Input(shape=(4096,))
pic = Dropout(rate=0.3)(pic_i)
pic = Dense(units=256,activation='relu')(pic)

both = add([seq,pic])
both = Dense(units=256,activation='relu')(both)
out = Dense(units=1754,activation='softmax')(both)

model = Model([pic_i,seq_i],[out])

model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
gen = Generator(tups_list,32)
model = tensorflow.keras.models.load_model('new_deal_v6.h5')
model.fit(gen,steps_per_epoch=200,epochs=30)
```
